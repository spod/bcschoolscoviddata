Leaderboards tracking which school, school district etc have the most exposures.

```python
import csv
import pandas as pd
import numpy as np
```

```python
exposures_df = pd.read_csv('data/bc_school_covid_events_raw_data.csv')
exposures_df = exposures_df.drop(columns=['Notification Date', 'City','Exposure Dates', 'Extra Info'])
exposures_df = exposures_df.assign(Exposures=1)
exposures_df = exposures_df.groupby(['School', 'School District', 'Health Region'], sort=True).count().nlargest(exposures_df.shape[0], columns='Exposures')
exposures_df.head(20)
```

```python
exposures_df.to_csv('data/leaderboard.csv')
```
